```python
import pickle

def storeData():
    # intializing the data 
    pickel_dataset_1 = ["mayur",100,200,300]
    pickel_dataset_2 = ("kadam","DYP","AKURDI")
    
    db = {}
    db['john'] = pickel_dataset_1
    db['jay'] = pickel_dataset_2
    dbfile = open("examplepickle", "wb")
    pickle.dump(db, dbfile)
    dbfile.close()

    
def loadData():
    # for reading
    
    dbfile = open("examplepickle", "rb")
    db = pickle.load(dbfile)
    for key in db:
        print(key, "==>", db[key])
    dbfile.close()

    
if __name__ =='__main__':
    storeData()
    loadData()
```
