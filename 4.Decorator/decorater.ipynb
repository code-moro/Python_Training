```python
#general Purpose Decorators

def dec_pass_args(function_dect):
    def wrapper_accepting_argument(*args,**kwargs):
        print("the positional argument are",args)
        print("the keyword argument are",kwargs)
        function_dect(*args)
    return wrapper_accepting_argument


@dec_pass_args
def funct_to():
    print("hello dec_To function")

funct_to()    
```
