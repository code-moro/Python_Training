```python
# To count number of digit in number in python
n =int(input("Enter Number"))
count=0
while(n>0):
    count=count+1
    n=n//10
print("The number of digits in the number are:",count)
```

```python
#chaining of decorator: we use functools library
from functools import wraps
#code for testing decorator chaining
def decor1(func):
    def inner():
        x=func()
        print("X in decor 2",x)
        return x*x    # logic doing mul
    return inner

def decor(func):
    def inner():
        x=func()
        print("X in decor first",x)
        return 2*x  #multiply with 2
    return inner
@decor1
@decor
def num():
    return 10

#calling num
num()
```
